```python
## Your code here
from functions import newton_root
p41 = Compose(Sin(), Exponential())
rt = newton_root(p41,1.0)
print(rt)
xmin, xmax = -2, 2
xs = np.linspace(xmin, xmax, 100)
plt.plot(xs, p41.evaluate(xs), 'k-', label = 'Sin(exp(x))') # plot the function
plt.hlines(0, xmin, xmax, colors='red', linestyles='dashed') # dashed line for y = 0
plt.scatter(rt, 0, c='k')
plt.text(rt + 0.05, -0.2, r'$root$') # root label
plt.xlim(xmin, xmax)
plt.ylim(xmin, xmax)
plt.legend()
plt.show()
```

## Part B

```python
## Your code here
from functions import newton_extremum
p51 = Compose(Sin(), Exponential())
extr = newton_extremum(p51,0.0)
print(extr)
xmin, xmax = -2, 2
xs = np.linspace(xmin, xmax, 100)
plt.plot(xs, p51.evaluate(xs), 'k-', label='Sin(exp(x))') # plot the function
plt.hlines(0, xmin, xmax, colors='red', linestyles='dashed') # dashed line for y = 0
plt.scatter(extr, p51.evaluate(extr), c='k')
plt.text(extr + 0.05, p51.evaluate(extr)-0.2, r'$extremum$') # extremum label
plt.xlim(xmin, xmax)
plt.ylim(xmin, xmax)
plt.legend()
plt.show()
```
